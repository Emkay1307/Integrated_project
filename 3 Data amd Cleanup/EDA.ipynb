```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
age_mean = insurance_data['age'].mean()
age_median = insurance_data['age'].median()
age_mode = stats.mode(insurance_data['age'])

# Display the mean, median, and mode of the 'age' column
print(f"Mean age: {age_mean}")
print(f"Median age: {age_median}")
print(f"Mode age: {age_mode.mode[0]} (occurs {age_mode.count[0]} times)")
```

```python
ge_mean = insurance_data['age'].mean()
age_median = insurance_data['age'].median()
age_mode = stats.mode(insurance_data['age']).mode[0]

# Create a histogram for the 'age' column
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.hist(insurance_data['age'], bins=20, edgecolor='black')  # Create a histogram with 20 bins

# Plot vertical lines for mean, median, and mode
plt.axvline(age_mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {age_mean:.2f}')
plt.axvline(age_median, color='green', linestyle='dashed', linewidth=1, label=f'Median: {age_median}')
plt.axvline(age_mode, color='blue', linestyle='dashed', linewidth=1, label=f'Mode: {age_mode}')

# Set labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age Distribution with Mean, Median, and Mode')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
